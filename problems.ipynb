```python
# importing required modules
import ast
import sys
import math

# this function asks the user to run the program agin or not if yes then it will run again the program else exit from program
def run_again():
    
    try:
        is_run = int(input("Do you want to run the program again enter 1 for yes or press any other key for exit : "))
        if is_run == 1:
            main()
        else:
            Exit_from_program()
    except:
        sys.exit()
    
        

# this function converts temperature from fahrenheit to celcius
def Convert_Degree_Fahrenheit_To_Degree_Celcius():
    
    temperature = float(input("Please Enter Temperature (in degree fahrenheit) : "))
    
    temperature_in_celcius = (5/9) * (temperature - 32)
    
    print("Temperature in degree celcius is : ", temperature_in_celcius)
    
    run_again()


#     this function converts temperature from celcius to fahrenheit
def Convert_Degree_Celcius_To_Degree_Fahrenheit():
    
    temperature = float(input("Please Enter Temperature (in degree celcius) : "))
    
    temperature_in_fahrenheit = (9/5) * temperature + 32
    
    print("Temperature in degree fahrenheit is : ", temperature_in_fahrenheit)
    
    run_again()
    
    
# this function calculate average marks and percentage
def Calculating_average_marks_and_percentage():
    
    max_marks = ast.literal_eval(input("Enter maximum marks of each subject in a form of tuple or simply seprated by comma : "))
    marks = ast.literal_eval(input("Enter marks obtained in each subject in a form of tuple or simply seprated by comma : "))
    
    if type(marks) == int or type(marks)==float:
        
        print("Average marks is : ","{:.2f}".format(marks))
        print("Percentage is : ","{:.2f}%".format(marks * 100 / max_marks))
        
    elif len(max_marks) != len(marks):
        raise Exception()   
        
    else:
        
        average_marks = sum(marks)/len(marks)
        percentage = sum(marks) * 100 / sum(max_marks)
        
        print("Average marks is : ","{:.2f}".format(average_marks))
        print("Percentage is : ","{:.2f}%".format(percentage))
        
    run_again()
        

# this function checks alphabet whether it is consonent or vowel        
def Check_alphabet_type():
    
    alphabet = input("Enter alphabet : ")
    
    if len(alphabet)==1 and ((ord(alphabet)>=65 and ord(alphabet)<=90) or (ord(alphabet)>=97 and ord(alphabet)<=122)):
        
        if (alphabet == "A" or alphabet == "E" or alphabet == "I" or alphabet == "O" or alphabet == "U" 
            or alphabet == "a" or alphabet == "e" or alphabet == "i" or alphabet == "o" or alphabet == "u"):
            print("{} is vowel".format(alphabet))
            
        else:
            print("{} is consonent".format(alphabet))
            
    else:
        raise Exception()
        
    run_again()
        
        
# checking number whether it is divisible by both 15 and 25
def Divisible_by_both_15_and_25():
    
    num = int(input("Enter Number : "))
    
    if num%15==0 and num%25==0:
        print("{} is divisible by both 15 and 25".format(num))
        
    else:
        print("{} is not divisible by both 15 and 25".format(num))
        
    run_again()
        

# this function calculates the electricity bill        
def Electricity_bill():
    
    for_100_units = 0
    from_101_to_200_units = 5
    after_200_units = 10
    
    units = float(input("Enter consumption units : "))
    
    bill = 0
    
    if units>100 and units <= 200:
        bill = (units-100) * from_101_to_200_units
        
    elif units>200:
        bill = (200-100)*from_101_to_200_units + (units-200)*after_200_units
        
    print("Your electricity bill is : ","Rs. {:.2f}".format(bill))
    
    run_again()
    

# this function takes car name as key and its price as values and stores in a dictinary and finally prints keys and values int the form of lists    
def Car_Models():
    
    car = {}
    flag = True
    
    while flag:
        
        car_name = input("Enter car name : ")
        car_price = float(input("Enter car price : "))
        
        car[car_name] = "$ {:.2f}".format(car_price)
        
        ask = int(input("Do you want to make another entry of car model if Yes enter 1 and 0 for NO : "))
        
        flag = True if ask==1 else False
        
    print("Dictionary Keys : ",list(car.keys()))
    print("Dictionary values : ",list(car.values()))
    
    run_again()
    
def half_pyramid_of_stars():
    
    rows = int(input("Enter number of rows you want to print in pyramid : "))
    
    if rows<=0:
        raise Exception()
        
    for i in range(rows):
        for j in range(i+1):
            print("*",end=" ")
        print()
        
    run_again()
    
def reverse_half_pyramid_of_stars():
    
    rows = int(input("Enter number of rows you want to print in pyramid : "))
    
    if rows<=0:
        raise Exception()
        
    for i in range(rows):
        for j in range(rows-i,0,-1):
            print("*",end=" ")
        print()
        
    run_again()
    
def half_pyramid_of_numbers():
    
    rows = int(input("Enter number of rows you want to print in pyramid : "))
    
    if rows<=0:
        raise Exception()
        
    for i in range(rows):
        for j in range(i+1):
            print(j+1,end=" ")
        print()
        
    run_again()
    
def reverse_half_pyramid_of_numbers():
    
    rows = int(input("Enter number of rows you want to print in pyramid : "))
    
    if rows<=0:
        raise Exception()
        
    for i in range(rows):
        n = 1
        for j in range(rows-i,0,-1):
            print(n,end=" ")
            n+=1
        print()
        
    run_again()
    
def half_pyramid_of_continous_alphabets():
    
    rows = int(input("Enter number of rows you want to print in pyramid : "))
    
    if rows<=0:
        raise Exception()
        
        
    for i in range(rows):
        a = 65
        for j in range(i+1):
            print(chr(a),end=" ")
            a+=1
        print()
        
    run_again()
    
def reverse_half_pyramid_of_continous_alphabets():
    
    rows = int(input("Enter number of rows you want to print in pyramid : "))
    
    if rows<=0:
        raise Exception()
        
    for i in range(rows):
        a = 65
        for j in range(rows-i,0,-1):
            print(chr(a),end=" ")
            a+=1
        print()
        
    run_again()
    

def half_pyramid_of_alphabets():
    
    rows = int(input("Enter number of rows you want to print in pyramid : "))
    
    if rows<=0:
        raise Exception()
        
    a = 65    
    
    for i in range(rows):
        
        for j in range(i+1):
            print(chr(a),end=" ")
        a+=1
        print()
        
    run_again()
    
def reverse_half_pyramid_of_alphabets():
    
    rows = int(input("Enter number of rows you want to print in pyramid : "))
    
    if rows<=0:
        raise Exception()
        
    a = 65 + rows - 1
    
    for i in range(rows):
        
        for j in range(rows-i,0,-1):
            print(chr(a),end=" ")
        a-=1
            
        print()
        
    run_again()
    
def half_pyramid_of_alphabets_with_continuity():
    
    rows = int(input("Enter number of rows you want to print in pyramid : "))
    
    if rows<=0:
        raise Exception()
        
    a = 65    
    
    for i in range(rows):
        
        for j in range(i+1):
            print(chr(a),end=" ")
            a+=1
        print()
        
    run_again()
    
def reverse_half_pyramid_of_alphabets_with_continuity():
    
    rows = int(input("Enter number of rows you want to print in pyramid : "))
    
    if rows<=0:
        raise Exception()
        
    a = 65
    
    for i in range(rows):
        
        for j in range(rows-i,0,-1):
            print(chr(a),end=" ")
            a+=1
            
        print()
        
    run_again()
    
def half_pyramid_of_continous_numbers():
    
    rows = int(input("Enter number of rows you want to print in pyramid : "))
    
    n = 1
    
    if rows<=0:
        raise Exception()
        
    for i in range(rows):
        for j in range(i+1):
            print(n,end=" ")
            n+=1
        print()
        
    run_again()
    
def reverse_half_pyramid_of_continous_numbers():
    
    rows = int(input("Enter number of rows you want to print in pyramid : "))
    
    if rows<=0:
        raise Exception()
        
    n = 1
        
    for i in range(rows):
        for j in range(rows-i,0,-1):
            print(n,end=" ")
            n+=1
        print()
        
    run_again()
    
def full_pyramid():
    
    rows = int(input("Enter number of rows you want to print in pyramid : "))
    
    if rows<=0:
        raise Exception()
        
        
    for i in range(rows):
        print(" "*(rows-i-1),end="")
        for j in range(i+1):
            print("*",end=" ")
        print()
        
    run_again()
    
def reverse_full_pyramid():
    
    rows = int(input("Enter number of rows you want to print in pyramid : "))
    
    if rows<=0:
        raise Exception()
        
        
    for i in range(rows):
        print(" "*(i),end="")
        for j in range(rows-i,0,-1):
            print("*",end=" ")
        print()
        
    run_again()
    
def pascal_traiangle():
    
    rows = int(input("Enter number of rows you want to print in pyramid : "))
    
    if rows<=0:
        raise Exception()
      
    l = [0]*rows
    l[0] = 1
    l1 = [0]*rows
    
    for i in range(rows):
        print(" "*(rows-i-1),end="")
        for j in range(i+1):
            print(l[j]+l[j-1],end=" ")
            l1[j] = l[j] + l[j-1]
        l = l1[:]
        print()
        
    run_again()
    

    
def is_prime(num):
    flag = True
    
    if num<2:
        return False
    
    for i in range(2 , int(math.sqrt(num))+1):
        if num%i==0:
            flag = False
            return False
        
    return flag
    
def printing_prime_numbers_in_a_range():
    
    start_num = int(input("Enter starting of range : "))
    end_num = int(input("Enter ending of range : "))
    
    if end_num<start_num or start_num<0:
        raise Exception()
        
    for i in range(start_num,end_num):
        if is_prime(i):
            print(i)
    
    run_again()
    
    
# this function exits from the program
def Exit_from_program():
    sys.exit()
    
# main function
def main():
    
    try:
    
        l = [Convert_Degree_Celcius_To_Degree_Fahrenheit , 
            Convert_Degree_Fahrenheit_To_Degree_Celcius , 
            Calculating_average_marks_and_percentage,
             Check_alphabet_type,
             Divisible_by_both_15_and_25,
             Electricity_bill,
             Car_Models,half_pyramid_of_stars,
             reverse_half_pyramid_of_stars,
             half_pyramid_of_numbers,
             reverse_half_pyramid_of_numbers,
             half_pyramid_of_continous_alphabets,
             reverse_half_pyramid_of_continous_alphabets,
             half_pyramid_of_alphabets,
             reverse_half_pyramid_of_alphabets,
             half_pyramid_of_continous_numbers,
             reverse_half_pyramid_of_continous_numbers,
             half_pyramid_of_alphabets_with_continuity,
             reverse_half_pyramid_of_alphabets_with_continuity,
             full_pyramid,
             reverse_full_pyramid,
             pascal_traiangle,
             printing_prime_numbers_in_a_range,
            Exit_from_program,
            ]

        print("List of Programs : ")
        print()

        for i , j in enumerate(l,1):
            print(str(i)+".  "+j.__name__.replace("_"," "))

        print()
        execute = int(input("Please enter the number of program which you want to execute : "))
        l[execute-1]()
        
    except KeyboardInterrupt:
        print("You stopped the program")
        
    except SystemExit:
        print("Exitted from program")
        
    except:
        
        print()
        print("Wrong input try again !")
        print()
        main()
    
    
if __name__ == "__main__":
    main()
    
```
